```python
def eval_crwl_ner(jsn, run_idx, short=True):
    if short:
        out=''.join(['|'+str(round(100*sum([0 if len(str(jsn[i]['entities'][0].get(z,'')))>0 
                                         else 1 for i in range(run_idx)])/(run_idx+1),2)) + '%' 
             for z in jsn[0]['entities'][0].keys()])

    return out
```

```python
import sys
def status_and_val_ner(dt0,run_idx,runs,nws_list,section):
    sys.stdout.write('\r'+str(run_idx+1) + '/'+str(runs) + ' ' +
                         nws_list[run_idx]['url']['Title'][:35] + 
                         " Progress {:2.2%}".format((run_idx +1)/ runs) + 
                         ''+ eval_crwl_ner(nws_list, run_idx) +
                         '|ETA: ' + str((runs-run_idx-1)*(datetime.datetime.now()-dt0)/(run_idx+1))[:4] + '/' +
                         str((runs)*(datetime.datetime.now()-dt0)/(run_idx+1))[:4] + 
                        '           ')
    sys.stdout.flush()
```

```python
import numpy as np
import json
import time
import datetime


#filename=np.random.choice(list_files)[:-4]
#filename='rss_crawl2019-04-25 23-00-26'

for filename in list_files:
    filename=filename[:-4]
    print('Running '+filename)
    path = './crwl/'
    path_out='./entities/'

    try:
        with open(path_out+filename+'_entities.txt') as f:
            news_repo = json.load(f)['key']    
        print('Continue ...')
    except:
        with open(path+filename+'.txt') as f:
            news_repo = json.load(f)['key']

    runs=len(news_repo)

    dt0=datetime.datetime.now()

    print_time('Begin')

    ii=0

    for run_idx in range(runs):
        try:
            news_repo[run_idx]['entities']
        except:
            qry=news_repo[run_idx]['url'].get('Title','') + ' ' + news_repo[run_idx]['url'].get('Content','')
            if qry!=' ':
                out=detect_entities(qry, language_client)

                res_ner=[{'name':out[idx].name,
                  'type':out[idx].type,
                  'salience':out[idx].salience,
                  'mid':out[idx].metadata['mid'],
                  'url':out[idx].metadata['wikipedia_url'],
                  'mentions_unq':list(set([out[idx].mentions[i].text.content for i in range(len(out[idx].mentions))])),
                  'mentions_len': len(out[idx].mentions)} 
                  for idx in range(len(out))]

                if res_ner==[]:
                    res_ner=[{None:None}]
            else:
                res_ner=[{None:None}]

            news_repo[run_idx]['entities']=res_ner

            if qry!=' ':
                status_and_val_ner(dt0,run_idx,runs,news_repo,'entities')

            ii=save_file(path_out+filename+'_entities.txt',news_repo,ii,run_idx,runs,freq=250)

    print_time('End')
```
